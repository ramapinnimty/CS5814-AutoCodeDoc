```python
# Importing all the necessary packages
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

# Java

```python
java_file  = open('java_out.txt', 'r')
java_lines = java_file.readlines()
```

```python
java_losses  = []
for line in java_lines:
    java_losses.append(float(line.strip()))
```

```python
plt.plot(java_losses)
plt.title("Loss vs Iteration for an epoch of Java")
plt.xlabel("Iteration")
plt.ylabel("loss")
```
